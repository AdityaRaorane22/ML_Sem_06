```python
import cv2
import mediapipe as mp

# Initialize Mediapipe Pose model
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5)

# Initialize OpenCV for video capture
cap = cv2.VideoCapture(0)

# Function to detect body language
def detect_body_language(landmarks):
    if landmarks:
        # 1. Detect Standing vs Sitting
        shoulder_left = landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER]
        shoulder_right = landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER]
        hip_left = landmarks[mp_pose.PoseLandmark.LEFT_HIP]
        hip_right = landmarks[mp_pose.PoseLandmark.RIGHT_HIP]
        
        if abs(shoulder_left.y - shoulder_right.y) < 0.05 and abs(hip_left.y - hip_right.y) < 0.05:
            print("Body Language: Standing Straight")
        else:
            print("Body Language: Sitting or Leaning Back")

        # 2. Detect Nodding (Up/Down Movement of the head)
        nose = landmarks[mp_pose.PoseLandmark.NOSE]
        left_eye = landmarks[mp_pose.PoseLandmark.LEFT_EYE]
        right_eye = landmarks[mp_pose.PoseLandmark.RIGHT_EYE]
        
        # Simple check for nodding (up or down)
        if nose.y < left_eye.y and nose.y < right_eye.y:
            print("Body Language: Nodding (Up)")
        elif nose.y > left_eye.y and nose.y > right_eye.y:
            print("Body Language: Nodding (Down)")
        
        # 3. Detect Eye Movement (Left/Right/Up/Down)
        if left_eye.x < right_eye.x:
            print("Body Language: Looking Left")
        elif left_eye.x > right_eye.x:
            print("Body Language: Looking Right")
        
        # Detect Eye Blink (Simple check)
        if left_eye.visibility < 0.5 and right_eye.visibility < 0.5:
            print("Body Language: Eye Blink Detected")

# Start video capture loop
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break

    # Flip the frame for later mirror display
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = pose.process(frame)

    # Convert the frame back to BGR for OpenCV display
    frame = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)

    # Draw pose landmarks on the frame
    if results.pose_landmarks:
        mp.solutions.drawing_utils.draw_landmarks(frame, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)
        
        # Call function to detect body language based on landmarks
        detect_body_language(results.pose_landmarks.landmark)

    # Display the frame
    cv2.imshow('Body Language Detection', frame)

    # Break the loop when 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp

# Initialize Mediapipe Pose model
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5)

# Initialize OpenCV for video capture
cap = cv2.VideoCapture(0)

# Function to detect and print body language
def detect_body_language(landmarks):
    # Example: Detect if the person is standing straight or sitting
    if landmarks:
        # Check the angle between key points to infer body language
        # A simple heuristic: If the angle between shoulder, elbow, and wrist is open, they might be raising their hands.
        # For more complex body language, more advanced checks are needed.

        # For demonstration, we assume "Straight posture" if shoulders and hips are aligned.
        shoulder_left = landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER]
        shoulder_right = landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER]
        hip_left = landmarks[mp_pose.PoseLandmark.LEFT_HIP]
        hip_right = landmarks[mp_pose.PoseLandmark.RIGHT_HIP]

        if abs(shoulder_left.y - shoulder_right.y) < 0.05 and abs(hip_left.y - hip_right.y) < 0.05:
            print("Body Language: Standing Straight")
        else:
            print("Body Language: Not Standing Straight")

# Start video capture loop
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break

    # Flip the frame for later mirror display
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = pose.process(frame)

    # Convert the frame back to BGR for OpenCV display
    frame = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)

    # Draw pose landmarks on the frame
    if results.pose_landmarks:
        mp.solutions.drawing_utils.draw_landmarks(frame, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)
        
        # Call function to detect body language based on landmarks
        detect_body_language(results.pose_landmarks.landmark)

    # Display the frame
    cv2.imshow('Body Language Detection', frame)

    # Break the loop when 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object
cap.release()
cv2.destroyAllWindows()
```
